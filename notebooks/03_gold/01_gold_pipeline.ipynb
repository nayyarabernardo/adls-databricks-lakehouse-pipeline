```python
dbutils.widgets.text("table_name", "")
table_name = dbutils.widgets.get("table_name")

if not table_name:
    raise ValueError("Parâmetro table_name é obrigatório")

from src.config.settings import LAYER_PATHS
from src.gold.setup.registry import GOLD_TABLES

if table_name not in GOLD_TABLES:
    raise ValueError(
        f"Tabela gold não suportada: {table_name}. "
        f"Disponíveis: {list(GOLD_TABLES.keys())}"
    )

cfg = GOLD_TABLES[table_name]

silver_path = LAYER_PATHS["silver"]
gold_path = f"{LAYER_PATHS['gold']}/{cfg['gold_suffix']}"

cfg["builder"](
    spark,
    silver_path,
    cfg["gold_table"],
    gold_path
)

print(f"gold {table_name} criada com sucesso")
```
