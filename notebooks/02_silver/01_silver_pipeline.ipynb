```python
dbutils.widgets.text("table_name", "")
table_name = dbutils.widgets.get("table_name")

if not table_name:
    raise ValueError("Parâmetro table_name é obrigatório")

from src.config.settings import LAYER_PATHS
from src.silver.setup.registry import SILVER_TABLES

if table_name not in SILVER_TABLES:
    raise ValueError(
        f"Tabela silver não suportada: {table_name}. "
        f"Disponíveis: {list(SILVER_TABLES.keys())}"
    )

cfg = SILVER_TABLES[table_name]

bronze_path = f"{LAYER_PATHS['bronze']}/{cfg['bronze_suffix']}"
silver_path = f"{LAYER_PATHS['silver']}/{cfg['silver_suffix']}"

cfg["builder"](
    spark,
    bronze_path,
    cfg["silver_table"],
    silver_path
)

print(f"Silver {table_name} criada com sucesso")
```
